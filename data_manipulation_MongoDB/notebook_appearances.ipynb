# Dropping the rows with NaN values

```python
df = df.dropna()
df
```

# Creating a function to filter only the europian characters

```python
def contains_only_european_chars(s):
    return re.match(r'^[a-zA-Z0-9\s\.,;:!?\-\'\"()€£\u00C0-\u017F]+$', s) is not None
```

# Changing the type of the columns

```python
df['player_name'] = df['player_name'].astype('str')
df['competition_id'] = df['competition_id'].astype('str')
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.dtypes
```

# Filtering the rows with the created function

```python
df_filtered = df[df['player_name'].apply(contains_only_european_chars)]
df_filtered = df[df['competition_id'].apply(contains_only_european_chars)]
```

```python
df_filtered
```

# Checking the dataset to match the clubs and the players datasets

```python
df_players = pd.read_csv('/home/lorenzo/Documents/3-TERZO-ANNO/IUMTWEB/Manipulated_data/players.csv')
df_clubs = pd.read_csv('/home/lorenzo/Documents/3-TERZO-ANNO/IUMTWEB/Manipulated_data/clubs.csv')
df_filtered_filtered = df_filtered[df_filtered['player_id'].isin(df_players['player_id'])]
df_filtered_filtered= df_filtered_filtered[df_filtered_filtered['player_current_club_id'].isin(df_clubs['club_id'])]
df_filtered_filtered
```
